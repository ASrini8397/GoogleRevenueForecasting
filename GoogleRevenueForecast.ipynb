```python
from prophet import Prophet
import pandas as pd 
import numpy as np
```

```python
df=pd.read_excel('/Users/apple/Desktop/revenuedata.xlsx', parse_dates=True)
```

```python
df.head()
```

```python
df.columns=['ds', 'y']
```

```python
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=275)
future.tail()
future
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```
